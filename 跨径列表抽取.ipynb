```python
import pandas as pd
import re
df = pd.read_excel("data/上部结构.xls")
def kuahao_process(x):
    result = []
    x = x.strip()
    match1 = re.match(r'2015-(\d+)-(\d+).*', x)#匹配样例；2015-1-14 00:00:00"
    match2 = re.match(r'(\d+)-(\d+)', x)#匹配样例18-29
    match3 = re.match(r'(\d+)',x)#匹配样例1,
    if match1:
        result.extend(list(range(int(match1.group(1)),int(match1.group(2))+1)))
        return result
    elif match2:
        result.extend(list(range(int(match2.group(1)),int(match2.group(2))+1)))
        return result
    elif match3:
        result.append(int(match3.group(1)))
        return result
    else:
        return result

    
def getKuahaolist(x):
    kuahao_unprocess_list = x.split(',')
    kuahao_process_list = []
    for kuahao in kuahao_unprocess_list:
        kuahao_process_list.extend(kuahao_process(kuahao))
    if len(kuahao_process_list)== 0:
        return 'nan'
    else:
        return  ','.join(str(v) for v in kuahao_process_list)

df['跨号列表'] = df['跨（孔）号'].astype(str).map(getKuahaolist)
df
# print(getKuahaolist("1-9,14-18"))
```
